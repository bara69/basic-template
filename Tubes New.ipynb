```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
path_to_csv = 'D:\Kuliah\S1\Python\Tubes.csv'
df = pd.read_csv(path_to_csv)
print("Columns:")
print(df.columns)
display(df)
```

```python
Kolesterol = df[df['chol'] >= 240]
Kolesterol
```

```python
Terjangkit = Kolesterol[Kolesterol['output']==1]
Terjangkit
```

```python
TdkTerjangkit = Kolesterol[Kolesterol['output']==0]
TdkTerjangkit
```

```python
plt.scatter(Terjangkit['age'], Terjangkit['chol'])
plt.scatter(TdkTerjangkit['age'], TdkTerjangkit['chol'])
plt.plot(Terjangkit['age'], Terjangkit['chol'])
plt.plot(TdkTerjangkit['age'], TdkTerjangkit['chol'])
plt.legend(['Terjangkit', 'TdkTerjangkit'])
plt.xticks(np.arange(min(TdkTerjangkit['age']), max(TdkTerjangkit['age']+1), 2.0))
plt.yticks(np.arange(min(Terjangkit['chol']), max(Terjangkit['chol']), 20.0))
plt.xlabel('Umur', fontsize=18)
plt.ylabel('Kolesterol', fontsize=18)
plt.tight_layout()
```

```python
plt.figure(figsize=(20, 20))
plt.barh(Terjangkit['age'], Terjangkit['chol'], 1.0)
plt.barh(TdkTerjangkit['age'], TdkTerjangkit['chol'], 1.0)
plt.yticks(np.arange(min(TdkTerjangkit['age']), max(TdkTerjangkit['age']+1)))
plt.xticks(np.arange(min(Terjangkit['chol']), max(Terjangkit['chol']+2), 10))
plt.xticks(rotation='vertical')
plt.title("Perbandingan Tingkat Kolesterol Penderita Penyakit Jantung dan Tidak")
bars1 = plt.barh(Terjangkit['age'], Terjangkit['chol'], color=sns.color_palette('dark'))
for bar in bars1:
    xval = bar.get_width()
    plt.text(xval, bar.get_y() + bar.get_height()/2, xval, ha='center', va='center')
bars2 = plt.barh(TdkTerjangkit['age'], TdkTerjangkit['chol'], color=sns.color_palette('bright'))
for bar in bars2:
    xval = bar.get_width()
    plt.text(xval, bar.get_y() + bar.get_height()/2, xval, ha='center', va='center')
plt.xlabel('Kolesterol', fontsize=18)
plt.ylabel('Umur', fontsize=18)
plt.tight_layout()
```
